# Classe com Método e propriedade

```python
class Horario:
    """Representa um horário do dia"""
    def __init__(self, hora=0, minuto=0, segundo=0):
        self.hora = hora
        self.minuto = minuto
        self.segundo = segundo
    
    def __str__(self):
        return f'{self.hora:02d}:{self.minuto:02d}:{self.segundo:02d}'
    
    def horario_to_inteiro(self):
        minutos = self.hora  * 60 + self.minuto
        segundos = minutos *  60  + self.segundo
        return segundos
    
    def inteiro_to_horario(segundos):
        minuto, segundo = divmod(segundos, 60)
        hora, minuto = divmod(minuto,  60)
        return Horario(hora, minuto, segundo)
    
    def __add__(self, outro):
        segundos = self.horario_to_inteiro() + outro.horario_to_inteiro()
        return Horario.inteiro_to_horario(segundos)
    
    def descrevaHorario(self):
        mensagem = f'Este objeto possui a hora {self.hora:02d}, '
        mensagem = mensagem + f'o minuto é {self.minuto:02d} '
        mensagem = mensagem + f'e o segundo é {self.segundo:02d}.'
        return mensagem
```

```python
natacao = Horario(20,15,0)
```

```python
natacao.descrevaHorario()
```

```python
natacao.__str__()
```

```python
print(natacao)
```

```python
inicio = Horario(20,20,0)
```

```python
duracao = Horario(3, 0, 0)
```

```python
fim = inicio + duracao
print(fim)
```

# Polimorfismo

```python
class Operacao:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def calcular(self):
        pass
```

```python
class Soma(Operacao):
    def calcular(self):
        return self.x + self.y
```

```python
total = Soma(10, 20)
total.calcular()
```

```python
class Multiplicacao(Operacao):
    def calcular(self):
        return self.x * self.y
```

```python
total = Multiplicacao(240, 3)
total.calcular()
```

```python
class Subtracao(Operacao):
    def calcular(self):
        return self.x - self.y

total = Subtracao(10, 20)
total.calcular()
```

# Herança

```python
class Pessoa:
    def __init__(self, nome, idade):
        self.nome = nome
        self.idade = idade
    
    def __str__(self):
        return f'Nome: {self.nome} - Idade: {self.idade}'

class Aluno(Pessoa):
    def __init__(self, nome, idade, matricula):
        super().__init__(nome, idade)
        self.matricula = matricula
    
    def __str__(self):
        return f'Nome: {self.nome} - Idade: {self.idade} - Matrícula: {self.matricula}'

class Professor(Pessoa):
    def nomeProfessor(self):
        return f'Professor: {self.nome}'

class Funcionario(Pessoa):
    def idadeFuncionario(self):
        return f'Idade do funcionário: {self.idade}'
```

```python
ana = Aluno('Ana', 20, '123456')
print(ana)	
```

```python
lucas = Professor('Lucas', 20)
```

```python
print(lucas)
```

```python
lucas.nomeProfessor()
```

```python
samara = Funcionario('Samara', 22)
```

```python
print(samara)
```

```python
samara.idadeFuncionario()
```
