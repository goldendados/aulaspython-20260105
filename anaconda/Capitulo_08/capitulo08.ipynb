# Gravando arquivos

```python
from os.path import basename, exists
def download(url):
    filename = basename(url)
    if not exists(filename):
        from urllib.request import urlretrieve
        local, _ = urlretrieve(url, filename)
        print("Downloaded " + str(local))
    return filename
```

```python
download("https://www.gutenberg.org/cache/epub/345/pg345.txt")
```

# Expressões Regulares (Regex)

```python
texto = "I am Dracula: and I bid you welcome, Mr. Harker, to my house."
```

```python
padrao = "Dracula"
```

```python
import re
```

```python
resultado = re.search(padrao, texto)
```

```python
type(resultado)
```

```python
resultado.string
```

```python
resultado.group()
```

```python
resultado.span()
```

```python
resultado2 = re.search("Jonathan", texto)
```

```python
type(resultado2)
```

```python
resultado2.span()
```

```python
def encontraPrimeiro(padrao):
    for linha in open("pg345_limpo_v2.txt"):
        resultado = re.search(padrao, linha)
        if resultado != None:
            return resultado
```

```python
resultado3 = encontraPrimeiro('Harker')
```

```python
resultado3.string
```

```python
padrao = 'Mina|Murray'
resultado4 = encontraPrimeiro(padrao)
resultado4.string
```

```python
def contaLinhas(padrao):
    contador = 0

    for linha in open("pg345_limpo_v2.txt"):
        resultado = re.search(padrao, linha)
        if resultado != None:
            contador += 1
    return contador
```

```python
resultado5 = contaLinhas(padrao)
print("Padrão regex: ", padrao, "Linhas:", resultado5)
```

```python
resultado6 = encontraPrimeiro('^Dracula')
resultado6.string
```

```python
resultado7 = encontraPrimeiro('Harker$')
resultado7.string
```

```python
resultado8  =  encontraPrimeiro('cent(er|re)')
resultado8.string
```

```python
resultado9 = encontraPrimeiro('colou?r')
resultado9.string
```

```python
linha = resultado9.string
linha
```

```python
re.sub('colou?r', 'color', linha)
```
