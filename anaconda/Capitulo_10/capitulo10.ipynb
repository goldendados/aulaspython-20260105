```python
listaPalavras = open('words.txt').read().split()
len(listaPalavras)
```

```python
''.join(reversed("Impacta"))
```

```python
def revertePalavra(palavra):
    return ''.join(reversed(palavra))
```

```python
# executando a consulta com lista
def metodoLento():
    contador = 0
    for palavra in listaPalavras:
        if revertePalavra(palavra) in listaPalavras:
            contador += 1
    return contador
```

```python
metodoLento()
```

```python
dictPalavras = {}
for palavra in listaPalavras:
    dictPalavras[palavra] = 1
```

```python
# executando a consulta com dicionario
def metodoRapido():
    contador = 0
    for palavra in dictPalavras:
        if revertePalavra(palavra) in dictPalavras:
            contador += 1
    return contador
```

```python
metodoRapido()
```

```python
def contaElementos(conjunto):
    contagem = {}
    for elemento in conjunto:
        if elemento not in contagem:
            contagem[elemento]=1
        else:
            contagem[elemento]+=1
    return contagem
```

```python
resultado = contaElementos("brontosaurus")
resultado
```

```python
for chave in resultado:
    valor = resultado[chave]
    print(chave,": ", valor, sep="")
```
