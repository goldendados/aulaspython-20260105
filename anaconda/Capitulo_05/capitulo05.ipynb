```python
x = 20
y = 10

# estrutura de decisão encadeada
if x > y:
    print("x é maior que y")
elif x < y:
    print("x é menor que y")
elif x == y:
    print("x é igual a y")
else:
    print("não há valores")
```

```python
x = 10 
y = 10

# estrutura de decisão aninhada
if x == y:
    print("x é igual a y")
else:
    if x < y:
        print("x é menor que y")
    else:
        print("x é maior que y")
```

```python
tipoDoProduto = "cozinha"
formaPagamento = "PIX"
```

```python
# sem aninhar
if tipoDoProduto == "cozinha" and formaPagamento == "PIX":
    print("Conceder desconto de 10%")
```

```python
# estilo aninhado
if tipoDoProduto == "cozinha":
    if formaPagamento == "PIX":
        print("Conceder desconto de 10%")
```

# Recursividade

```python
# sem recursividade
def contageRegressiva(num):
    for i in range(num):
        print(num-i)
    print("Fim")
```

```python
contageRegressiva(5)
```

```python
# com recursividade
def contagemRegressiva(num):
    if num <= 0:
        print("Fim")
    else:
        print(num)
        contagemRegressiva(num-1)
```

```python
contagemRegressiva(5)
```
