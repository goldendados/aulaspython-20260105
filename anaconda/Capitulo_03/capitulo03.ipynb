# Parâmetros

```python
def printDuasVezes(texto):
    print(texto)
    print(texto)
```

```python
printDuasVezes("Impacta")
```

```python
mensagem = "Olá"
```

```python
printDuasVezes(mensagem)
```

```python
def repete(palavra, n):
    print(palavra * n)
```

```python
repete("Oi,", 3)
```

```python
repete('Spam, ', 4)
```

```python
def repeteDuasVezes(texto):
    repete(texto, 4)
    print("Lovely Spam, Wonderful Spam!")
    repete(texto, 4)
```

```python
repeteDuasVezes("Spam, ")
```

# Repetição (For)

```python
for i in range(5):
    print(i)
```

```python
print("10 x 1 = ", 10  * 1)
print("10 x 2 = ", 10  * 2)
print("10 x 3 = ", 10  * 3)
print("10 x 4 = ", 10  * 4)
print("10 x 5 = ", 10  * 5)
print("10 x 6 = ", 10  * 6)
print("10 x 7 = ", 10  * 7)
print("10 x 8 = ", 10  * 8)
print("10 x 9 = ", 10  * 9)
print("10 x 10 = ", 10  * 10)
```

```python
for numero in range(1, 11):
    print("10 x ", numero, " = ", 10  * numero) 
```

```python
def printTabuada(num, linhas):
    for numero in range(1, linhas+1):
        mensagem = str(num) + " x " + str(numero) + " = "
        resultado = num * numero
        print(mensagem, resultado)
```

```python
printTabuada(5, 5)
```

```python
printTabuada(3, 40)
```
