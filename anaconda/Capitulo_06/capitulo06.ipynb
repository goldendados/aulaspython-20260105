```python
def minhaAbs(valor):
    if valor < 0:
        return -valor
    else:
        return valor
```

```python
minhaAbs(-250)
```

```python
minhaAbs(-250) * 2
```

# Funções booleanas

```python
def ehDivisivel(x, y):
    if x % y == 0:
        return True
    else:
        return False
```

```python
ehDivisivel(6 , 4)
```

```python
ehDivisivel(10, 5)
```

```python
if ehDivisivel(10,5):
    print("É divisivel")
```

```python
if ehDivisivel(10,3):
    print("É divisivel")
else:
    print("Não é divisivel")
```
