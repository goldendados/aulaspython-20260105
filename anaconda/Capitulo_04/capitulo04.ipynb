```python
def quadrado(tamanho):
    for i in range(4):
        forward(tamanho)
        left(90)
```

```python
make_turtle()

quadrado(30)
quadrado(60)
```

```python
def polygon(n, tamanho):
    angulo = 360 / n
    for i in range(n):
        forward(tamanho)
        left(angulo)
```

```python
make_turtle()
polygon(7, 30)
```

```python
import math

def circle(raio):
    circunferencia = 2 * math.pi *  raio
    n  = 30
    tamanho = circunferencia  / n
    polygon(n, tamanho)
```

```python
make_turtle(delay=0.02)
circle(30)
```

```python
def polyline(n, tamanho, angulo):
    for i in range(n):
        forward(tamanho)
        left(angulo)
```

```python
def polygon(n, tamanho):
    angulo = 360 / n
    polyline(n, tamanho, angulo)
```

```python
def arc(raio, angulo):
    arc_tamanho = 2 * math.pi * raio * angulo / 360
    n = 30
    tamanho = arc_tamanho / n
    passos_angulo =  angulo / n
    polyline(n, tamanho, passos_angulo)
```

```python
def circle(raio):
    arc(raio, 360)
```

```python
make_turtle(delay=0)
polygon(20, 9)
arc(70, 70)
circle(10)
```

# Docstring

```python
def tabuada(num, linhas):
    """Esta função realiza o calculo da tabuada
    a partir de um numero informado e a quantidade
    de linhas desejadas"""
    
    for i in range(1,linhas+1):
        print(num  * i)
```

```python
tabuada(10,  4)
```
