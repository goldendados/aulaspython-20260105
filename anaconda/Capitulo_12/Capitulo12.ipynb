```python
from os.path import basename, exists
def download(url):
    filename = basename(url)
    if not exists(filename):
        from urllib.request import urlretrieve
        local, _ = urlretrieve(url, filename)
        print("Downloaded " + str(local))
    return filename
```

```python
def linhaEspecial(linha):
    return linha.startswith("***")
```

```python
linhaEspecial("ola")
```

```python
linhaEspecial("***ola")
```

```python
def arquivoLimpo(arquivo_original, arquivo_limpo):
    leitura = open(arquivo_original,encoding='utf-8')
    escrita = open(arquivo_limpo,"w")

    for linha in leitura:
        if linhaEspecial(linha):
            break

    for linha in leitura:
        if linhaEspecial(linha):
            break
        escrita.write(linha)

    leitura.close()
    escrita.close()
```
