```python
from os.path import basename, exists
def download(url):
    filename = basename(url)
    if not exists(filename):
        from urllib.request import urlretrieve
        local, _ = urlretrieve(url, filename)
        print("Downloaded " + str(local))
    return filename
```

```python
def verificaEquiva(caminho1, caminho2,  caminho3):
    arq1 = open(caminho1, 'rb').read()
    arq2 = open(caminho2, 'rb').read()
    arq3 = open(caminho3, 'rb').read()
    return (arq1 == arq2) or (arq1 == arq3)
```

```python
verificaEquiva(arquivo1,arquivo2,arquivo3)
```

```python
verificaEquiva(arquivo1,arquivo3,arquivo2)
```

```python
verificaEquiva(arquivo3,arquivo1,arquivo2)
```

```python
def gerarHexa(arquivo):
    dados = open(arquivo,'rb').read()
    md5 = hashlib.md5()
    md5.update(dados)
    hexa = md5.hexdigest()
    return hexa
```
